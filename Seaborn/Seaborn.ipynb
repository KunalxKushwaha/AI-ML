# Seaborn:
Data Visualization Library.

 Importing the Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as pd
import pandas as pd
```

Note: [Seaborn has some build in Dataset]

```python
# Insert a theme for the Plots
sns.set_theme()
```

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Visualization the data
sns.relplot(data=tips, x = 'total_bill', y = 'tip', col='time', hue='smoker', style='smoker', size = 'size')
```

```python
# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

Scatter Plot

```python
sns.scatterplot(x ='sepal_length', y = 'petal_length',hue='species',data=iris)
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=iris)
```

```python
# Loading the titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.shape
```

Count Plot

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='survived', data=titanic)
```

Bar Chart

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
# House Price Dataset
from sklearn.datasets import fetch_california_housing
house_california =fetch_california_housing()

house = pd.DataFrame(house_california.data, columns = house_california.feature_names)
house['PRICE'] = house_california.target
```

```python
print(house_california)
```

```python
house.head()
```

Distribution Plot

```python
sns.distplot(house['PRICE'])
```

Correlation Matrix:
- Positive Correlation
- Negative Correlation

Heat Map

```python
correlation = house.corr()
```

```python
# Construcuting heat Map
plt.figure(figsize = (10,10))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.2f', annot = True, annot_kws={'size':8}, cmap='Blues')
```
